```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
from IPython.display import Audio

f1 = 200
f2 = 20
f = 20000
```

```python
fig, ax = plt.subplots()
t = np.arange(0,1,1/f)
y = np.cos(2*np.pi*f1*t) * np.cos(2*np.pi*f2*t)
ax.plot(np.arange(len(y)),y)
ax.set_ylabel('|Amplitude|')
ax.set_xlim(0,f/4)
```
