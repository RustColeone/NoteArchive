```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(0,40,41)
s_n = 2 * x * np.power(np.e, -0.1 * x)
z = np.zeros(41)
rand_uniform = np.random.rand(41,1)
z = np.reshape((1*(rand_uniform>0.5) - 0.5),len(z))

print(x, end = "\n\n")
print(s_n, end = "\n\n")
print(z, end = "\n\n")

fig,ax = plt.subplots()
ax.plot(x,s_n)
```

```python
x_n = s_n + z
fig,ax = plt.subplots()
ax.plot(x,x_n)
```

```python
fig,ax = plt.subplots()
ax.plot(x,s_n)
ax.plot(x,x_n)
```

```python
def FivePointFilter(ToFilter):
    y = np.zeros(41 + 5 - 1)
    N = len(ToFilter)
    for i in range(N): 
        for j in range(5):
            if (i-j>=0 and i-j<N): 
                y[i] = y[i] + ToFilter[i-j]/5;
    return y

y_n = FivePointFilter(x_n)
x1 = np.linspace(0,44,45)
fig,ax = plt.subplots()
ax.plot(x1,y_n)
ax.plot(x,x_n)
```

```python
z_n = FivePointFilter(z)

fig,ax = plt.subplots()
ax.plot(x1,z_n)
ax.plot(x,z)
```

```python
print("The rapid peak appears to be smoothened and changes less rapidly")
```
