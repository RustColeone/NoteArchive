```python
import numpy as np
from jupyterthemes import jtplot
jtplot.style()
# allows matplotlib charts to be displayed inside the notebook
%matplotlib inline  
import matplotlib.pyplot as plt
```

```python
fs = 5000               # sampling frequency (Hz)
t_max = 0.2             # signal duration (sec)
```

```python
N = round(t_max*fs)     # number of samples
t = np.linspace(0,t_max,N)
s_t = 10 + 20*np.cos(200*np.pi*t+np.pi/4) + 10*np.cos(500*np.pi*t)  
```

```python
# Plot signal in time domain
fig, ax = plt.subplots()
ax.plot(t,s_t)
ax.set_xlabel("t (sec)")
ax.set_xlim(0,t_max)       # use this command to zoom in the x-axis if needed
```

```python
S_f = np.fft.fft(s_t)             # list of FFT coefficients
freqs = fs/N*np.linspace(0,N-1,N) # list of frequencies corresponding to the coefficients returned by fft
```

```python
# Plot magnitude of FFT coefficients 
fig, ax = plt.subplots()
ax.plot(freqs,abs(S_f))
ax.set_xlabel("f (Hz)")
ax.set_xlim(0,fs)           # you might need to adjust the value here to better see the dominant frequencies
print('The sampling frequency is', fs, 'Hz')
```

```python
# Plot magnitude of FFT coefficients 
fig, ax1 = plt.subplots()
ax1.plot(freqs,abs(S_f))
ax1.set_xlabel("f (Hz)")
ax1.set_xlim(0,fs/10)           # you might need to adjust the value here to better see the dominant frequencies
print('The sampling frequency is', fs, 'Hz')
```

```python
def SpectrumFind(Fs):
    fs = Fs                # sampling frequency (Hz)
    t_max = 0.2             # signal duration (sec)
    N = round(t_max*fs)     # number of samples
    t = np.linspace(0,t_max,N)
    s_t = 10 + 20*np.cos(200*np.pi*t+np.pi/4) + 10*np.cos(500*np.pi*t)
    S_f = np.fft.fft(s_t)             # list of FFT coefficients
    S_f = np.fft.fftshift(S_f)
    print(len(S_f))
    freqs = fs/N*np.linspace(0,N-1,N) # list of frequencies corresponding to the coefficients returned by fft
    # Plot magnitude of FFT coefficients 
    
    freqs_centered = fs/N * np.linspace(-N/2 +1, N/2, N)
    
    fig, ax = plt.subplots()
    ax.plot(freqs_centered,abs(S_f))
    ax.set_xlabel("f (Hz)")
    ax.set_xlim(-fs/2,fs/2)           # you might need to adjust the value here to better see the dominant frequencies
    print('The sampling frequency is', fs, 'Hz')
SpectrumFind(700)
```
