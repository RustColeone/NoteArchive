## 1.0 Why 2D-arrays / matrices?

In the previous labs, we saw that one-dimensional (1D) arrays are useful to represent 1D discrete signals. Specifically, we used the array structure of the Numpy package and we refered to the 1D arrays as **vectors**.

Signal-processing engineers are not only interested in 1D signals. For example, a variety of SP techniques are motivated by the need to store, process, transform and generate **images**. Images are two-dimensional (2D) signals. While acoustic signals show variation over the time-domain, images vary over (usually) *two spatial coordinates*.

Recall that we represent an accoustic signal with an one-dimensional continuous-time function $s(t), -\infty<t<+\infty$, and, it's sampled discrete version with an 1D sequence or **vector** $s[n],n\in\mathbb{Z}$. 

Similarly, we represent a 2D-image with a **2D-array** or **matrix** 
$
I[m,n],
$
where the two spatial coordinates $m$ and $n$ take integer values. Every pair of coordinates $(m,n)$ represents the location of a single **image pixel**.

<font color='red'>**Exercise 1.2.4**</font> Write a Python function that takes as inputs a matrix A, two indices i and j, and a target value val. The function sets 
the value of the entry $A[i,j]$ equal to the target value.

```python
def set_specific_entry(A,i,j,val):
    dim = A.shape
    rows = dim[0]
    cols = dim[1]
    if (i >= rows and j >= cols) or (i < 0 and j < 0):
        return "i and j out of range"
    elif i >= rows or i < 0:
        return "i out of range"
    elif j >= cols or j < 0:
        return "j out of range"
    try:
        A[i][j] = float(val)
    except:
        return "Val is not a valid number"
    return A

print(set_specific_entry(A,3,4,99))
print(set_specific_entry(A,2,3,98))
```

<font color='red'>**Exercise 1.2.6**</font>  Write a Python function that takes as input a matrix $A$ and does the following. For every entry of the matrix, set the value to zero if pixel value is less than 0.5, otherwise change the pixel value to one. Call the matrix with the matrix A given to you as input.

```python
def threshold_mtx(M):
    dim = M.shape
    rows = dim[0]
    cols = dim[1]
    for i in range(rows):
        for j in range(cols):
            x = M[i][j]
            if(x < 0.5):
                M[i][j] = 0
            else:
                M[i][j] = 1
    return M

R = np.random.rand(4,3)
print(R)

R_thresholded = threshold_mtx(R)
print(R_thresholded)
```

#### The transpose of a matrix

The **transpose** of an $m \times n$ matrix $A$ is a new matrix $B$ of dimensions $n \times m$, such that
$$
B[j][i] = A[i][j],\quad \text{for all } i,j
$$
Usually we denote the transpose of $A$ by $A^T$. 

We will learn more about transposes in the lectures. For now, let us experiment with them using Python.

```python
# To find the transpose of A
B = A.T

print("The transpose matrix is ", B)
```
